```python
import wikipedia
import re
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
wiki = wikipedia.page("Python")
text = wiki.content
text = re.sub(r'==.*?==+', '', text)
text = text.replace('\n', ' ')
```

```python
def plot_cloud(word_cloud):
    plt.figure(figsize=(40,30))
    plt.imshow(word_cloud)
    plt.axis("off")
```

```python
wordcloud = WordCloud(width=3000, height=2000, random_state=1, background_color='salmon', colormap="plasma", stopwords=STOPWORDS, collocations=False).generate(text)
plot_cloud(wordcloud)
```
